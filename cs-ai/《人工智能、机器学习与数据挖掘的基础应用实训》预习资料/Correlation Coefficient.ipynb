```python
import numpy as np
from matplotlib import pyplot as plt
```

## Positively linearly correlated

```python
X = np.random.uniform(-1,1,100)
Y = X
corrcoef = np.corrcoef(X, Y)

ax = plt.gca()
ax.set_aspect(1)
plt.scatter(X, Y, s=0.1)
plt.text(0, -0.5, s=f'corrcoef: {corrcoef[0][1]:.4f}')
plt.show()
```

## Negatively linearly correlated

```python
X = np.random.uniform(-1,1,100)
Y = -X
corrcoef = np.corrcoef(X, Y)

ax = plt.gca()
ax.set_aspect(1)
plt.scatter(X, Y, s=0.1)
plt.text(0, -0.5, s=f'corrcoef: {corrcoef[0][1]:.4f}')
plt.show()
```

## Independence $\rightarrow$ zero correlation

```python
X = np.random.uniform(-1,1,100)
Y = np.random.uniform(-1,1,100)
corrcoef = np.corrcoef(X, Y)

ax = plt.gca()
ax.set_aspect(1)
plt.scatter(X, Y, s=0.1)
plt.text(0, -0.5, s=f'corrcoef: {corrcoef[0][1]:.4f}')
plt.show()
```

## zero correlation $\not \rightarrow$ independence

```python
thetas = np.random.uniform(0,2 * np.pi,100)
X = np.sin(thetas)
Y = np.cos(thetas)
corrcoef = np.corrcoef(X, Y)

ax = plt.gca()
ax.set_aspect(1)
plt.scatter(X, Y, s=0.1)
plt.text(0, -0.5, s=f'corrcoef: {corrcoef[0][1]:.4f}')
plt.show()
```

$p(Y) \ne p(Y|X=x), \forall x$

## Linearly correlated in "second order"

```python
thetas = np.random.uniform(0,2 * np.pi,100)
X, Y = np.sin(thetas), np.cos(thetas)
XX, YY = X * X, Y * Y
corrcoef = np.corrcoef(XX, YY)

ax = plt.gca()
ax.set_aspect(1)
plt.scatter(X, Y, s=0.1)
plt.text(0, 0.2, s=f'corrcoef: {corrcoef[0][1]:.4f}')
plt.show()
```
