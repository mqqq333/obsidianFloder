```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Data Preparation

$$
z = x^2 + y^2
$$

```python
x = np.arange(-5,5,0.5)
y = np.arange(-5,5,0.5)
X, Y = np.meshgrid(x, y)
Z = X ** 2 + Y ** 2

ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap='rainbow')
ax.set_xlabel('$x$', fontsize=20)
ax.set_ylabel('$y$', fontsize=20)
ax.set_zlabel('$z$', fontsize=20)
plt.show()
```

```python
zipped = zip(X.flatten() ** 2, Y.flatten() ** 2, Z.flatten())
table = pd.DataFrame(zipped, columns=['x^2','y^2','z'])
print(table.head())
data = table.to_numpy()
```

# Plotting Cost-Param Graph

$$
z' = [x^2, y^2] \cdot [u, v]^T \\
C = \text{MSE}_\text{cost} = \mathrm{E}[(z - z')^2]
$$

```python
u = np.arange(-2, 4, 0.5)
v = np.arange(-2, 4, 0.5)
U, V = np.meshgrid(u, v)
W = np.vstack((U.reshape(1,-1), V.reshape(1,-1)))
Zp, Z = data[:, :2] @ W, data[:, 2:]
C = ((Zp - Z) ** 2).mean(axis=0)

ax = plt.axes(projection='3d')
ax.plot_surface(U,V,C.reshape(12,12),cmap='rainbow')
ax.set_xlabel('$u$', fontsize=20)
ax.set_ylabel('$v$', fontsize=20)
ax.set_zlabel('$C$', fontsize=20)
plt.show()
```
